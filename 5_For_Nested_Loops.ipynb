#### 1. While Loop In Python

While Loop In Python is used to execute a block of statement as long as a given condition is true. And when the condition is false, the control will come out of the loop.

The condition is checked every time at the beginning of the loop.

```python
##### While Loop Syntax

while (condition):
    # statements
```

image.png

The condition is a boolean expression that is evaluated before each iteration of the loop. If the condition is True, the code block inside the loop is executed. If the condition is False, the loop is terminated and control is passed to the next statement after the loop.

In the backend, a while loop is implemented using a combination of a jump table and a conditional jump instruction. When the loop starts, the interpreter creates a jump table that stores the address of the instruction that follows the loop. The interpreter then evaluates the condition and, if it is True, executes the code block inside the loop. After the code block is executed, the interpreter jumps back to the instruction in the jump table and evaluates the condition again. This process is repeated until the condition is False, at which point the interpreter moves on to the next statement after the loop.

While loops are useful when you want to repeat a block of code an unknown number of times, such as when you are waiting for user input or when you are processing a large dataset. They are also useful for performing tasks that require a certain condition to be met, such as searching for a value in a list or checking for errors in a program.

It is important to make sure that the condition in a while loop eventually becomes False, otherwise the loop will run indefinitely and could cause your program to crash. You can use the break statement to exit a while loop early if a certain condition is met, or you can use the continue statement to skip the rest of the code block and go back to the beginning of the loop.

```python
x = 0
while(x<5):
    print(x)
    x=x+1
```

#### While Loop With Else In Python
The else part is executed if the condition in the while loop becomes False.

```python
x = 1
while (x < 5):
   print('inside while loop value of x is ',x)
   x = x + 1
else:
   print('inside else value of x is ', x)
```

#### 2. For Loop In Python

A for loop in Python is a looping construct that allows you to iterate over a sequence of elements, such as a list or a string. It allows you to execute a block of code repeatedly for each element in the sequence.

```python
# syntax
for val in sequence:
    # statements
```

image.png

In the backend, a for loop in Python is implemented using an iterator. An iterator is an object that allows you to iterate over a sequence of elements. When you use a for loop, Python creates an iterator for the sequence you are looping over and calls the ( ___next_ __) method on the iterator to retrieve each element in the sequence.

For example, when you use a for loop to iterate over a list, Python creates a list iterator object that keeps track of the current position in the list and returns each element in the list as you loop over it.

There are several reasons why you might use a for loop in Python:

* To perform a specific task for each element in a sequence. For example, you might use a for loop to print each element in a list or to perform some calculations on each element in a list.

* To repeat a task multiple times. For example, you might use a for loop to print a message 10 times.

* To iterate over a sequence of elements in a list, tuple, string, or other iterable object. For example, you might use a for loop to iterate over a list of numbers and add them together to find the sum.

Overall, for loops are a useful and powerful tool for iterating over sequences of elements in Python. They allow you to perform tasks repeatedly and make your code more efficient and easier to read.

```python
for i in range(1,5):
    print(i)
```

##### The range() Function In Python
The range() function is a built-in that is used to iterate over a sequence of numbers.

* Syntax Of range() Function:
> range(start, stop[, step])

##### The range() Function Parameters
* start: Starting number of the sequence.
* stop: Generate numbers up to, but not including this number.
* step(Optional): Determines the increment between each numbers in the sequence.

```python
# Example 1 of range() function
for i in range(10):
    print(i)
```

```python
# Example 2 of range() function
for i in range(2,9):
    print(i)
```

```python
# Example 3 of range() function using step parameter
for i in range(2,9,2):
    print(i)
```

```python
# Example 4 of range() function
for i in range(0,-10,-2):
    print(i)
```

```python
# Example to iterate a list using for loop
my_list = [42, 61, 47, 46, 91, 18]
for i in my_list:
    print(i)
```

### 3. Nested For loops

Nested for is nothing but a for loop inside another for a loop.



```python
rows = 5
for i in range(1, rows + 1):
    for j in range(1, i + 1):
        print("*", end=" ")
    print('')
```

```python
for i in range(1,6):
    for j in range(0,i):
        print(i,end=" ")
    print('')
```

```python
for i in range(1,6):
    for j in range(5,i-1,-1):
        print(i,end=" ")
    print('')
```

```python
n = 5
for i in range(1,n+1):
    print (((" ") *(n-i))+((str(i)+ " ")*i))
```
